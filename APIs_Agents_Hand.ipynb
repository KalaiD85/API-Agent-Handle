Task 1 — Fetch and Validate the Response

```python
import requests

def fetch_countries():
    url = "https://restcountries.com/v3.1/all?fields=name,capital,region"
    try:
        response = requests.get(url, timeout=10)  # timeout for safety
        print("HTTP Status Code:", response.status_code)

        if response.status_code == 200:
            countries = response.json()
            print("Total number of countries retrieved:", len(countries))
        else:
            print(f"Request failed with status code {response.status_code}")
    except requests.exceptions.RequestException as e:
        print("An error occurred while fetching data:", e)

# Run the function
fetch_countries()
```

Task 2 — Extract and Display Key Fields

```python
import requests
import pandas as pd

# Fetch data from RestCountries API
url = "https://restcountries.com/v3.1/all?fields=name,population,region,area"
response = requests.get(url)

if response.status_code == 200:
    countries = response.json()

    # Extract required fields
    data = []
    for country in countries:
        data.append({
            "common_name": country["name"]["common"],
            "population": country.get("population", None),
            "region": country.get("region", None),
            "area": country.get("area", None)
        })

    # Create DataFrame
    df = pd.DataFrame(data)

    # Display first 10 rows
    print(df.head(10))
else:
    print("Failed to fetch data. Status code:", response.status_code)
```

Task 3 - Summarize the Data

```python
# Group by region and sum population
region_population = df.groupby("region")["population"].sum()

# Find region with highest total population
highest_population_region = region_population.idxmax()
highest_population_value = region_population.max()

# Find country with largest area
largest_area_country = df.loc[df["area"].idxmax(), "common_name"]
largest_area_value = df["area"].max()

# Print results clearly
print("Region with Highest Total Population:", highest_population_region, "-", highest_population_value)
print("Country with Largest Area:", largest_area_country, "-", largest_area_value)
```
